### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
num = int(input("enter a num:"))
total=0

for i in range(1, num + 1):
    if i % 2 == 0:
        total -= i
    else:
        total += i

print("ans is:",total)
```

### 13) WAP to print sum of digits of given number.

```python
num = int(input("enter a number to sum"))
digit,total = 0,0

while num !=0:
    digit = num%10
    total += digit
    num //=10

print("the sum of digit is:",total)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
num = int(input("enter a number to check Palindrome"))
rev,rem = 0,0

a = num

while num !=0:
    digit = num%10
    rev = rev*10+digit
    num //=10

if(a == rev):
    print("the",a,"is palindrome:")
else:
    print("the",a,"is  not palindrome:")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
num =int(input("enter a num to check armstrong"))
digit,count,sum=0,0,0

a=num

while a !=0:
    digit = a%10
    count +=1
    a//=10

a=num

while a!=0:
    digit =a%10
    sum += digit**count
    a//10

if num == sum:
    print("the num is armstrong")
else:
    print("the num is not armstrong")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n= int(input("enter a number enter a list of perfect num"))
num =0
for num in range(1,n+1):
    sum=0
    for i in range(1,num):
        if num%i == 0:
            sum +=i

    if sum == num) and (num!=0):
        print(num)
```
