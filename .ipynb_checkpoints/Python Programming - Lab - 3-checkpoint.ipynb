### 03) WAP to find out largest number from given two numbers using simple if and ternary operator.

```python
getnum1 = int(input("Enter the num1:"))
getnum2 = int(input("Enter the num2:"))

if(getnum1 > getnum2):
    print("the given",getnum1,"is maximum")
else:
    print("the given",getnum2,"is maximum")

res = getnum1 if getnum1 > getnum2 else getnum2
print("the given",res,"is maximum")
```

### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("entetr the year"))
if (year%4 == 0 and year%100!=0)or (year%400 == 0):
    print("the",year,"is leap year")
```

### 06) WAP in python to display the name of the day according to the number given by the user.

```python
days = int(input("enter the num of the day:(1-7)"))

match days:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("invalid number")
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
getnum1 = int(input("Enter the num1:"))
getnum2 = int(input("Enter the num2:"))
choice = int(input ("give choice 1 for +,2 for -,3 for *,4 for /\n"))
match choice:
    case 1:
        res = getnum1 + getnum2
        print("res is :",res)
    case 2:
         res = getnum1 - getnum2
         print("res is :",res)
    case 3:
         res = getnum1 * getnum2
         print("res is :",res)
    case 4:
         res = getnum1 / getnum2
         print("res is :",res)
    case _:
        print("invalid choice")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
s1,s2,s3,s4,s5,*s=map(int,input("give 5 subjects marks seprate by ,").split(","))

percent = float(s1+s2+s3+s4+s5)/5

if percent >= 70:
    print("Distinction pass :",percent,"%")
elif percent >= 60 and percent < 70:
    print("First Class :",percent,"%")
elif percent >= 45 and percent < 60:
    print("Second Class:",percent,"%")
elif percent >= 35 and percent < 45:
    print("Pass Class :",percent,"%")
else:
    print("Fail  :",percent,"%")
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
unit = int(input("enter the electricity units"))

bill = 0
if unit<= 50:
    bill = unit*2.60
elif unit<= 100:
    bill = (50*2.60)+(unit-50)*3.25
elif unit<=200:
    bill = (50*2.60)+(50*3.25)+(unit-100)*5.26
else:
    bill=(50*2.60)+(50*3.25)+(100*5.26)+(unit-200)*8.45
print(F"Total Electricity in Rs:{bill:f}")
```
