### 13) WAP to find second largest number in a list.

```python
li = list(map(int, input().split()))

max1 =0
max2 =0

for i in li:
        if i > max1:
            max2 = max1
            max1 = i
        elif i > max2 and i != max1:
            max2 = i

print(max2,max1)
```
